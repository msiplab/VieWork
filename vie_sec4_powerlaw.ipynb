# べき乗測変換

\begin{equation}
  s = T(r) = (L-1)\left(\frac{r}{L-1}\right)^\gamma
\end{equation}

### URLを指定してアクセスする方法
- 参照サイト：　http://homepages.cae.wisc.edu/~ece533/images/

```
import urllib.request
import io

url = 'http://homepages.cae.wisc.edu/~ece533/images/lena.png'
file_name = io.BytesIO(urllib.request.urlopen(url).read())
```

## 画像の読み込み
画像処理ライブラリ Scikit-image を利用

```
from skimage import io
img = io.imread(file_name)
img.dtype
```

## 画像の表示
可視化ライブラリMatplotlibを利用

```
from matplotlib import pyplot as plt

plt.imshow(img)
plt.axis('off')
```

## グレースケール画像に変換

```
from skimage.color import rgb2gray

gray_img = rgb2gray(img)
display(gray_img.dtype)
gray_img.max()
```

## グレースケール画像の表示
PIL画像をNumPy配列(float32)に変換し，[0,1]の範囲に正規化して表示

```
import numpy as np

plt.imshow(gray_img, cmap='gray')
plt.axis('off')
```

## べき乗測変換のグラフプロット
パラメータ$\gamma$を設定して関数をプロット

```
gamma = 0.5

r = np.linspace(0,1,100)
s = np.power(r,gamma) 
plt.plot(r,s)
plt.axis('square')
plt.xlabel('r')
plt.ylabel('s')
plt.show()
```

## 画像のべき乗測変換
NumPyライブラリの power 関数を利用

```
res_img = np.power(gray_img,gamma)
```

## 処理結果の表示

```
plt.imshow(res_img,cmap='gray')
plt.axis('off')
```

# 符号なし整数8ビットに変換


```
res_img_uint8 = np.uint8(255.0*res_img)
display(res_img_uint8.dtype)
res_img_uint8.max()
```
