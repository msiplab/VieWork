# べき乗測変換

\begin{equation}
  s = T(r) = (L-1)\left(\frac{r}{L-1}\right)^\gamma
\end{equation}

## 画像ファイルのアップロード

```
from google.colab import files
uploaded_file = files.upload()
```

```
uploaded_file_name = next(iter(uploaded_file))
print(uploaded_file_name)
```

## 画像の読み込み
画像処理ライブラリ Pillow (PIL)を利用

```
from PIL import Image

img = Image.open(uploaded_file_name)
print(img.mode)
print(img.format)
print(img.size)
print(img.getbands())
```

## 画像の表示
可視化ライブラリMatplotlibを利用

```
from matplotlib import pyplot as plt

plt.imshow(img)
plt.axis('off')
```

## グレースケール画像に変換

```
gray_img = img.convert('L')
from matplotlib import pyplot as plt
print(gray_img.mode)
print(gray_img.format)
print(gray_img.size)
print(gray_img.getbands())
```

## グレースケール画像の表示
PIL画像をNumPy配列に変換して表示

```
import numpy as np

gray_array = np.asarray(gray_img)
plt.imshow(gray_array, cmap='gray')
plt.axis('off')
```

## べき乗測変換のグラフプロット
パラメータ$\gamma$を設定して関数をプロット

```
gamma = 0.5

r = np.linspace(0,1,100)
s = np.power(r,gamma) 
plt.plot(r,s)
plt.axis('square')
plt.xlabel('r')
plt.ylabel('s')
plt.show()
```

## 画像のべき乗測変換
NumPyライブラリの power 関数を利用

```
res_array = np.power(gray_array,gamma)
```

## 処理結果の表示

```
plt.imshow(res_array,cmap='gray')
plt.axis('off')
```
